# **2. Data Cleaning:**

Initial Inspection: The dataset

*   Initial Inspection: The dataset comprises 20,052 entries and 680 columns.

*   Missing Values: Many columns have missing values, with some exceeding 50%. These columns were removed, and the remaining missing values were filled with the mean for numerical columns.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Check for missing values
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values_percentage = (missing_values / len(df)) * 100

missing_values
missing_values_percentage
```

```python
# Filter columns with more than 50% missing values
cols_to_drop = missing_values[missing_values_percentage > 50].index
```

```python
# Drop columns with more than 50% missing values
df_cleaned = df.drop(cols_to_drop, axis=1)
df_cleaned
```

```python
# Fill remaining missing values with the mean for numerical columns
for col in df_cleaned.select_dtypes(include=['float64']).columns:
    df_cleaned[col].fillna(df_cleaned[col].mean(), inplace=True)
```
